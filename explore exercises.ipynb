```python
import pandas as pd
import numpy as np
import acquire
import prepare as p 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = p.prepare_store_data()
```

```python
#verify that everything pulled in properly 
df.head(1)
```

```python
train = df[:'2016'] # includes 2016
test = df['2017']

plt.plot(train.index, train.total_price)
plt.plot(test.index, test.total_price)
plt.show()
```

```python
# gonna store our target variable in a y variable
y = train.total_price
y.plot.hist()
```

```python
y.resample('M').sum().plot()
```

```python
y.resample('M').sum().sort_index().plot()
```

```python
y.plot(alpha=.2, label='Hourly')
y.resample('D').mean().plot(alpha=.5, label='Daily')
y.resample('W').mean().plot(alpha=.8, label='Weekly')
y.resample('M').mean().plot(label='Montly')
y.resample('Y').mean().plot(label='Yearly')
plt.legend()
plt.show()
```

```python
y.plot(alpha=.2, label='Hourly')
y.resample('D').mean().rolling(3).mean().plot(alpha=.5, label='3 day')
y.resample('D').mean().rolling(7).mean().plot(alpha=.8, label='1 week')
plt.legend()
```
